##  Define a function that differentiates between continuous and discrete variables. 
 Hint: A number of unique values might be useful. Store continuous data into a continuous variable and do the same for discrete and categorical.
 


```python
# know the unique values for each column
for c in data.select_dtypes(np.number).columns.tolist():         
    print(c, len(data[c].unique()))
```

```python

def cont_disc(arg):
    uniquevalues = len(arg.unique())
    if uniquevalues > 100:
        x = ['continuous',arg.name]
        return x
    else:
        x = ['discrete',arg.name]
        return x
```
